```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd
```

```python
url = 'https://results.golfstat.com/public/leaderboards/gsnav.cfm?pg=stats&tid=2875&statsType=c'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html')
table = soup.find_all('div')
```

```python
body_tag = soup.body
```

```python
#by round by hole idx = 1018

#859
#898
#937
#976 gap:39
chart = []
for idx,child in enumerate(body_tag.descendants):
    #print('-----------')
    #print('idx:',idx)
    #print(child)
    #print('------------')
    rows = list(range(313,977,39))
    #print(rows)
    if idx in rows:
        chart.append(child.text)
```

```python

df = pd.DataFrame(columns= ['hole', 'rank', 'par', 'yard', 'average', 'dif', 'eagle', 'birdie', 'par',
                            'bogie', 'dbl_bogie', 'other'])

for row in chart:

    df.loc[len(df)] = row.split('\n')[1:13]
        
```
