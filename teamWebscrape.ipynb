```python
pars = soup.find_all('div', attrs={"style":"display:inline-block;background-color:#05643A;color:white;text-align:center;border:1px solid silver;border-collapse:collapse;width:2em;"})

```

```python
headings = [par.get_text() for par in pars]
```

```python
site = soup.find_all('div')
```

```python
for i in range(0,len(site)):
    print('-'*100)
    print('\n')
    print(i)
    print('\n'*5)
    print(site[i])

```

```python
players = site[4].get_text().split('\n')[133:184]
```

```python
players
```

```python
values = []
```

```python
for player in players:
    if player == "":
        continue
    else:
        values.append(player)
```

```python
values
```

```python
elements = [x for x in values if values.index(x) in [1,5,9,13,17,21]]
```

```python
lst = []
```

```python
for element in elements:
    lst.append(list(map(int, element)))
```

```python
#9,10,20-end
```

```python
for l in lst:
    indexes = [9, 10, 20,21,22,23]
    for index in sorted(indexes, reverse=True):
        del l[index]
```

```python
lst
```

```python
headings[36:]
```

```python
sewanneRd1 = pd.DataFrame(lst)
```

```python
sewanneRd1 = sewanneRd1.T
```

```python
sewanneRd1['par'] = headings[36:]
```

```python
sewanneRd1
```

```python
sewanneRd1['averagePar'] = (sewanneRd1[0] + sewanneRd1[1] + sewanneRd1[2] + sewanneRd1[3] + sewanneRd1[4] + sewanneRd1[5]) / 6
```

```python
sewanneRd1
```

```python
sewanneRd1['diff'] = sewanneRd1['averagePar'] - sewanneRd1['par'].astype(float)
```

```python
sewanneRd1
```

```python
holes = list(range(1,19))
```

```python
sewanneRd1['Hole'] = holes
```

```python
sewanneRd1['par'] = sewanneRd1['par'].astype(int)
```

```python
sewanneRd1.set_index('Hole')
```

```python
sns.set(font_scale=1.5)
sewanneRd1.plot(x="Hole", y=['par', 'averagePar'], kind="bar", figsize=(15,8))
plt.ylabel('Par')
plt.title('Course Par vs Sewanne Par (Round 1)')
plt.legend(['CoursePar', 'AvgPlayerPar'])
#plt.savefig('sewVScourse.png')
```

```python
plt.figure(figsize=(15,8))
sns.set(font_scale=1.5)
sns.barplot(x=holes , y=sewanneRd1['diff'], color='blue')
plt.ylabel('Over/Under Par')
plt.xlabel('Hole')
plt.title('Sewanne Avg Par (Round 1)');
#plt.savefig('sewAvgPar.png')
```
