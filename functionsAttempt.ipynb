```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def courseStats(link):
    url = f'{link}'
    resp = requests.get(url).text
    soup = BeautifulSoup(resp)

    headings = soup.find_all('div', attrs={"style":"display:inline-block; border:1px solid silver;text-align:center;width:2.5em;background-color:#05643A;color:#FFF;"})
    headings_list = [i.get_text() for i in headings][:11]
    headings_list.insert(5, 'SC')

    scores = soup.find_all('div', attrs={"style":"clear:both;;white-space:nowrap;"})
    score_rows = [i.get_text() for i in scores]
    score_list = [score.split('\n')[1:13] for score in score_rows]

    df = pd.DataFrame(score_list, columns= headings_list)
    df = df.astype(float)
    return df
```

```python
stoneBridge = courseStats('https://results.golfstat.com/public/leaderboards/gsnav.cfm?pg=stats&tid=23540&statsType=c')
```

```python
def getPageSource(url, team, rd, driver_path):
    options = webdriver.ChromeOptions()
    options.add_argument('--ignore-certificate-errors')
    options.add_argument('--incognito')
    options.add_argument('--headless')
    driver = webdriver.Chrome(f"{driver_path}", options=options)

    driver.get(f"{url}")

    time.sleep(1)
    click_button = driver.find_element_by_link_text(f'{team}')
    click_button.click()

    time.sleep(3)
    click_round1 = driver.find_element_by_link_text(f'{rd}')
    click_round1.click()
    time.sleep(1)
    click_again = driver.find_element_by_link_text(f'{rd}')
    click_again.click()

    return driver.page_source
```

```python
page_source = getPageSource(staturl, 'Emory (B)', '1', path)
```

```python
page_sourcerd2 = getPageSource(staturl, 'Emory (B)', '2', path)
```

```python
def playerStats(pg_s, dataf):
    sp = BeautifulSoup(pg_s)
    site = sp.find_all('div')
    
    players = site[4].get_text().split('\n')[133:175]
    values = []
    for player in players:
        if player == "":
            continue
        else:
            values.append(player)
    
    parHeadings = dataf[:18].sort_values(by=['HOL'])['PAR'].iloc[:, 0]

    playerScores = [x for x in values if values.index(x) in [1,5,9,13,17,21]]
    scoresList = []

    for score in playerScores:
        scoresList.append(list(map(int, score)))

    for score in scoresList:
        indexes = [9, 10, 20,21,22,23]
        for index in sorted(indexes, reverse=True):
            del score[index]

    scoresList.append(parHeadings)
    colNames = ['p1', 'p2', 'p3', 'p4', 'p5', 'par']
    playerDict = dict(zip(colNames, scoresList))

    return pd.DataFrame(playerDict)
```

```python
rd1 = playerStats(page_source, stoneBridge)
```

```python
rd2 = playerStats(page_sourcerd2, stoneBridge)
```

```python
def setupDF(frame):
    frame['hole'] = [i for i in range(1,19)]
    frame['par'] = frame['par'].astype(int)
    frame['averagePar'] = frame[['p1', 'p2', 'p3', 'p4', 'p5']].sum(axis=1)/5
    frame['diff'] = frame['averagePar'] - frame['par']
    
    return frame
```

```python
rd1df = setupDF(rd1)
```

```python
rd2df = setupDF(rd2)
```

```python
finaldf = pd.concat([rd1df, rd2df])
```
